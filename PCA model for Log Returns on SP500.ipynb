Fit a PCA model to log returns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the data
data = pd.read_csv("SP500_log_returns.csv")

# Assuming the log returns are columns in your DataFrame, you can extract the values
X = data.values

# Standardize the data
mean = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)
X_standardized = (X - mean) / std_dev

# Fit a PCA model
pca = PCA(n_components=10)
principal_components = pca.fit_transform(X_standardized)

# Calculate the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the scree plot
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.8, align='center')
plt.title('Scree Plot')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()

print(pca.explained_variance_ratio_)
```

Scree Plot Includes Cumulative Variance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the data
data = pd.read_csv("SP500_log_returns.csv")

# Assuming the log returns are columns in your DataFrame, you can extract the values
X = data.values

# Standardize the data
mean = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)
X_standardized = (X - mean) / std_dev

# Fit a PCA model
pca = PCA(n_components=129)
principal_components = pca.fit_transform(X_standardized)

# Calculate the cumulative explained variance ratio
cumulative_explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative variance plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance_ratio) + 1), cumulative_explained_variance_ratio, marker='o', linestyle='-', color='b')
plt.title('Cumulative Variance Retained by Top 129 Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()

print(cumulative_explained_variance_ratio)
```

Plot the Weights of the 1st Principal Component by Industry Sector

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load log returns data
log_returns_data = pd.read_csv("SP500_log_returns.csv")

# Load ticker data
ticker_data = pd.read_csv("SP500_ticker_(filtered).csv")

# Extract 'numeric' columns for PCA
numeric_columns = log_returns_data.select_dtypes(include=np.number).columns
X = log_returns_data[numeric_columns].values

# Standardize the data
mean = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)
X_standardized = (X - mean) / std_dev

# Fit a PCA model
pca = PCA()
principal_components = pca.fit_transform(X_standardized)

# Get the weights for the 1st principal component
weights_1st_component = pca.components_[0]

# Create a DataFrame with ticker and weights
weights_df = pd.DataFrame({"ticker": log_returns_data.columns[1:], "weight_1st_component": weights_1st_component})

# Merge with sector information
merged_df = pd.merge(weights_df, ticker_data[["ticker", "sector"]], on="ticker")

# Plot the weights grouped by sector
plt.figure(figsize=(12, 8))
for sector, group in merged_df.groupby("sector"):
    plt.scatter(group["ticker"], group["weight_1st_component"], label=sector, alpha=0.7)

plt.title("Weights of 1st Principal Component Grouped by Sector")
plt.xlabel("Ticker")
plt.ylabel("Weight of 1st Principal Component")
plt.xticks(rotation=90)
plt.legend()
plt.show()

# Create a DataFrame with tickers and their corresponding sectors
ticker_sector = ticker_data[['ticker', 'sector']]

# Add a column for the weights of the 1st principal component
ticker_sector['weight_1st_component'] = weights_1st_component

# Group by sector and calculate the mean weight for each sector
mean_weights_by_sector = ticker_sector.groupby('sector')['weight_1st_component'].mean()

# Plot the bar plot
plt.figure(figsize=(10, 6))
mean_weights_by_sector.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Mean Weight of 1st Principal Component per Sector')
plt.xlabel('Sector')
plt.ylabel('Mean Weight')
plt.xticks(rotation=45, ha='right')
plt.show()
```
